```python
import json 
import pandas as pd
   
with open("../file_metadata.json", "r") as f:
    metadata = json.load(f)   
```

```python
short_metadata = {}
for src in metadata:
    short_metadata[src] = {k : v for k,v in metadata[src].items() if k not in ['raw_metadata', 'imagej_metadata']}
    
df = pd.DataFrame(short_metadata).T 
df
```
